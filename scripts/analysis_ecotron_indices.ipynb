```python
import pandas as pd
import os

# Folder path
folder_path = '../climpact_indices/indices/'

# File names
file_names = [
    'daily_ecotron_data_for_climpact_txx_ANN.csv',
    'daily_ecotron_data_for_climpact_tnn_ANN.csv',
    'daily_ecotron_data_for_climpact_fd_ANN.csv',
    'daily_ecotron_data_for_climpact_su_ANN.csv',
    'daily_ecotron_data_for_climpact_gsl_ANN.csv',
    'daily_ecotron_data_for_climpact_prcptot_ANN.csv',
    'daily_ecotron_data_for_climpact_rx1day_ANN.csv',
    'daily_ecotron_data_for_climpact_r10mm_ANN.csv',
    'daily_ecotron_data_for_climpact_cdd_ANN.csv',
    'daily_ecotron_data_for_climpact_cwd_ANN.csv'
]

# Initialize an empty DataFrame for combined data
combined_data = pd.DataFrame()

# Load and process each file
for file_name in file_names:
    # Construct the full file path
    file_path = os.path.join(folder_path, file_name)

    # Load the data, skipping the first 6 rows and excluding the last column
    data = pd.read_csv(file_path, skiprows=6, usecols=[0, 1])

    # Rename columns for consistency
    data.columns = ['date', file_name.split('_')[5]]  # Extracts the variable name from the file name

    # Exclude rows with missing values
    data.dropna(inplace=True)

    # Merge with the combined data
    if combined_data.empty:
        combined_data = data
    else:
        combined_data = pd.merge(combined_data, data, on='date', how='outer')

# Now, combined_data contains all the variables with their respective dates

# Drop rows with NaN values in any column
# Exclude rows where 'su' is -99.9
combined_data_cleaned = combined_data[combined_data['su'] != -99.9]
```

```python
# Let's calculate the average for each specified 3-year interval
# First, we'll define the intervals
year_intervals = [(1951, 1953), (2011, 2013), (2028, 2030), (2043, 2045), (2067, 2069), (2091, 2093)]

# Now we'll create a new DataFrame to hold the averages for these intervals
interval_averages = pd.DataFrame(columns=combined_data_cleaned.columns[2:])  # Exclude 'Unnamed: 0' and 'date'

# Process each interval
for start_year, end_year in year_intervals:
    # Filter the data for the current interval
    interval_data = combined_data_cleaned[(combined_data_cleaned['date'] >= start_year) & (combined_data_cleaned['date'] <= end_year)]
    
    # Calculate the mean for the interval and append it to the interval_averages DataFrame
    mean_values = interval_data.mean().to_dict()
    mean_values['year'] = f"{start_year}-{end_year}"  # Use a string to represent the interval
    interval_averages = interval_averages.append(mean_values, ignore_index=True)

# Since the year column is at the end, we'll move it to the front for readability
cols = list(interval_averages.columns)
cols = [cols[-1]] + cols[:-1]
interval_averages = interval_averages[cols]

# Display the averages for each interval
interval_averages
```

```python
# Re-importing pandas as the execution state was reset
import pandas as pd

# Re-calculate the interval averages as the state was reset
# Define the intervals
year_intervals = [(1951, 1953), (2011, 2013), (2028, 2030), (2043, 2045), (2067, 2069), (2091, 2093)]
interval_averages = pd.DataFrame(columns=combined_data_cleaned.columns[2:])  # Exclude 'Unnamed: 0' and 'date'

# Process each interval
for start_year, end_year in year_intervals:
    interval_data = combined_data_cleaned[(combined_data_cleaned['date'] >= start_year) & (combined_data_cleaned['date'] <= end_year)]
    mean_values = interval_data.mean().to_dict()
    mean_values['year'] = f"{start_year}-{end_year}"  # Use a string to represent the interval
    interval_averages = interval_averages.append(mean_values, ignore_index=True)

# Set 'year' as the index and drop the unnecessary columns
interval_averages.set_index('year', inplace=True)
interval_averages.drop(columns=['date'], inplace=True)

# Transpose the DataFrame
transposed_interval_averages = interval_averages.transpose()

# Specify the desired order of the rows
desired_order = ['txx', 'tnn', 'fd', 'su', 'gsl', 'prcptot', 'rx1day', 'r10mm', 'cdd', 'cwd']

# Reorder the rows according to the desired order
transposed_interval_averages = transposed_interval_averages.loc[desired_order]

# To round the values to 2 significant digits, we can apply the round function
rounded_interval_averages = transposed_interval_averages.round(2)

rounded_interval_averages
```
